```python
 

import pandas as pd
import numpy as np
import seaborn as sns; sns.set_style("darkgrid")
from sklearn import linear_model
import random
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# Levanto los datos

df = pd.read_csv('n0to20_100capMochila_10pesoMaximo.csv')
```

```python

axis1 = df.plot(x='n',logy=True)
plt.ylabel(u'Tiempo de ejecuci√≥n (mlseg).')
plt.xlabel('Cantidad de pedidos a procesar')
x1,x2,y1,y2 = axis1.axis()
plt.show()
```

```python
plt.clf()
df2 = pd.read_csv('n0to80_100capMochila_10pesoMaximo.csv')
df4 = pd.read_csv('n0to20_capMochila0to20000_10pesoMaximo.csv')

axis1 = plt.subplot(1,2,1)
plt.ylabel('Tiempos en milisegundos')
axis2 = plt.subplot(1,2,2)
df2.plot(x='n', ax=axis1, logy=True)
df4.plot(x='W', ax=axis2)
axis1.set_xlabel('(a) Cantidad de pedidos')
axis2.set_xlabel('(b) Capacidad de la mochila')

plt.show()
```
